```python
from matplotlib import pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster 
from sklearn.cluster import KMeans 
import pandas as pd 
from sklearn.cluster import AgglomerativeClustering

dftrain = pd.read_csv('C:/Users/tszch/Desktop/cs3481/vertebral_column_data/column_3C.dat', header= None, sep = ' ')  
X = dftrain.values[:, 0:6] 
Y = dftrain.values[:, 6]   
training_labels = dftrain.iloc  [:, -1]
#plt.plot(X[0,0],X[:,1] )

 
from sklearn import preprocessing 
le   = preprocessing.LabelEncoder() 
label = le.fit_transform(training_labels)
print(label)
```

```python
ml=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
ml.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=ml.fit_predict(X))
```

```python
Z = linkage(X,'complete')
plt.figure(figsize=(25, 10))
dendrogram(Z) 
plt.show() 
```

```python
kclusters = fcluster(Z,3, criterion='maxclust') 
kclusters

```

```python
import numpy as np
kclusters = np.where(kclusters== 1, 0, kclusters) 
kclusters = np.where(kclusters== 2, 1, kclusters) 
kclusters =np.where(kclusters== 3, 2, kclusters) 
kclusters
```

```python
km = KMeans(n_clusters=3) 
km.fit(X)
km.labels_ 
 
from sklearn.metrics import accuracy_score
accuracy_score(km.labels_, label)
```

```python
accuracy_score(kclusters, label)
```

```python
Z = linkage(X,'single')
plt.figure(figsize=(25, 10))
dendrogram(Z) 
plt.show() 
kclusters = fcluster(Z,3, criterion='maxclust') 
kclusters 
```

```python
kclusters = np.where(kclusters== 1, 0, kclusters) 
kclusters = np.where(kclusters== 2, 1, kclusters) 
kclusters =np.where(kclusters== 3, 2, kclusters) 
accuracy_score(kclusters, label)
```

```python
Z = linkage(X,'average')
plt.figure(figsize=(25, 10))
dendrogram(Z) 
plt.show() 
kclusters = fcluster(Z,3, criterion='maxclust') 
kclusters 
```
